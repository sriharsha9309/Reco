```python
import pandas as pd
from ast import literal_eval
```

```python
df=pd.read_csv(r"G:\movies_metadata.csv\movies_metadata.csv")
```

```python
df.head()
```

```python
df['genres'] = df['genres'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
df.head()
```

```python
vote_counts=df['vote_count']
m=vote_counts.quantile(0.95)
m
```

```python
vote_avg=df['vote_average']
c=vote_avg.mean()
```

```python
final=df[(df['vote_count'] >= m) & (df['vote_count'].notnull()) & (df['vote_average'].notnull())]
```

```python
final=final.reindex(columns=['title', 'year', 'vote_count', 'vote_average', 'popularity', 'genres'])
```

```python
final=final.drop('year',axis=True)
```

```python
final.head()
```

```python
def weighted_rating(x):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R)
```

```python
final['wr'] = final.apply(weighted_rating, axis=1)
```

```python
final= final.sort_values('wr', ascending=False).head(250)
```

```python
final.head()
```

```python
s = final.apply(lambda x: pd.Series(x['genres']),axis=1).stack().reset_index(level=1, drop=True)
```

```python
s.name='genres'
```

```python
final = final.drop('genres', axis=1).join(s)
```

```python
final
```

```python
def build_chart(generes):
    new=[]
    for i in range(1,15):
        if generes == df['generes'][i] and wr >5:
            new.append(df['title'][i])
            return new
```

```python
def reco(genre):
    matching_movies = [title for title in  final['genres'] == genre]
    a=final[matching_movies].sort_values('wr', ascending=False).head(11)
    return a['title']
```
